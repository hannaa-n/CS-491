```python
import transformers
from transformers import pipeline
```

## Text Generation

```python
generator = pipeline('text-generation', model='gpt2')
generated_text = generator("The meaning of life is", max_length=50, num_return_sequences=3)
# Adjust max_length and num_return_sequences as needed
for text in generated_text:
    print(text['generated_text'])
```

## Summarization

```python
summarizer = pipeline("summarization", model="google/flan-t5-base")
input_text = """Artificial intelligence (AI) is rapidly transforming various aspects of our lives,
from how we work and communicate to how we access information and make decisions.
AI systems are being used in healthcare to diagnose diseases and develop new treatments,
in finance to detect fraud and manage investments, and in transportation to power self-driving cars.
While the potential benefits of AI are immense, there are also concerns about its impact on jobs,
privacy, and security. It is crucial to develop AI responsibly and ethically to ensure that
it benefits humanity as a whole."""  # Replace with your desired text
summary = summarizer(input_text, min_length=5, max_length=50)
print(summary[0]['summary_text'])
```

## Translation
I choose to translate from english to vietnamese. After reading the translated vietnamese, I would say the accuracy is ~85%

```python
translator = pipeline("translation_en_to_vi", model="Helsinki-NLP/opus-mt-en-vi")
translated_text = translator("This is a test translation. Python is a popular programming language widely used in data science, artificial intelligence (AI), natural language processing (NLP), and more. Much of the content in these fields is presented using Jupyter Notebooks, an open-source, interactive web application that allows you to create and share documents containing live code, equations, visualizations, and explanatory text. While primarily used with Python, Jupyter Notebooks also support other programming languages.")
print(translated_text[0]['translation_text'])
```
